# 5. **Pandas DataFrames:**<br><center>Data Aggregation, Grouping, & Operations

> <font color=DeepPink>Using the `numpy` random data generator, create a dataframe with the following columns: Weather, Food, Price, and Number.

```python
# Import libraries for data aggregation (random data generator)
import numpy as np
from numpy.random import seed, rand, randint

# Create DF with specified columns
## 'Number' column must include randomn integers between 1 and 9
seed(45)
df_five = pd.DataFrame({'Weather':['cold', 'cold', 'hot', 'cold', 'hot', 'hot', 'cold', 
                                   'cold'], 
                        'Food':['ice cream', 'ice cream', 'soup', 'soup', 'ice cream', 'soup',
                                'ice cream', 'soup'],
                        'Price': 10 * rand(8), 
                        'Number': randint(1, 9, size=8)})
# NOTE: randint() must include the size parameter to generate different values
## Otherwise, the same random value will be assigned to each variable
### Source: NumPy.org - https://numpy.org/doc/stable/reference/random/generated/
### numpy.random.randint.html

# View DataFrame
df_five
```

> **NOTE:** The data will have four(4) columns:
>> * `Weather` & `Food` - strings
>> * `Price` - random float
>> * `Number` - a random integer between one(1) and nine(9)

> **NOTE:** Here is some info on pseudo random numbers vs true randomness, NumPy's random.seed() method, BitGenerators, and Generators.
>> __[What Does `random.seed` Do in NumPy](https://towardsdatascience.com/random-seed-numpy-786cf7876a5f)__

> <font color=DeepPink>**Group** the data by the weather column and then **create** a function to iterate through the groups <br>(You should have 2 groups after, for hot and cold)

```python
# Group the data by 'Weather' column ('weather_group')
## .groupby() method returns a groupby object which contains info about the groups.
## When a series is passed as an argument, the series values determine the groups.
### Source: pandas - https://pandas.pydata.org/pandas-docs/stable/reference/api/
### pandas.DataFrame.groupby.html
weather_group = df_five.groupby('Weather')

# Create a function to iterate through the groups
## Use 'for' loop because: 1. it iterates a fixed number of times
## 2. 'for' loops are always used with iterable objects, in this case the groupby object
## Source: wiki.python - https://wiki.python.org/moin/ForLoop

i = 0
# i isn't a hard-coded variable, you can use any name you like; think of it as i = iteration.
for name, group in weather_group:
    # 'name' represents the group name and 'group' represents the actual grouped DF
    # Source: geeksforgeeks - https://www.geeksforgeeks.org/
    # how-to-iterate-over-dataframe-groups-in-python-pandas/
    i = i + 1
    print("Group", i, name)
    print(group)
```

> <font color=DeepPink>Your function/variable that you created (weather_group) can be used for aggregation methods
>> * <font color=DeepPink>**Print** the ***first row, last row,*** and ***mean*** for each group

```python
# Aggregation functions return the values of multiple rows as a grouped, single summary value
# Source:SparkBy-https://sparkbyexamples.com/pandas/pandas-aggregate-functions-with-examples/

print("First rows of weather groups: \n", weather_group.first())
print("\nLast rows of weather groups: \n", weather_group.last())
print("\nMean of the values for each weather group:\n", weather_group.mean(numeric_only=True))
# NOTE: DataFrame.mean() was updated in version 2.0.0 from what is printed in the textbook
# Only using .mean() will raise a TypeError because the new version defaults to False, 
# meaning it can work on only float, int, and boolean columns and is not implemented for
# Series. To avoid this error, set the numeric_only argument to True.
# Source: pandas - https://pandas.pydata.org/docs/reference/api/
# pandas.DataFrame.mean.html#pandas.DataFrame.mean
```

> <font color=DeepPink>**Create** another group - on Food (so you would have Weather and Food)

```python
# Create another group, this time grouped on multiple columns ('Weather' & 'Food')
weather_food = df_five.groupby(['Weather', 'Food'])
print("Combinations of weather and food values as groups:\n", weather_food.groups)
# Lists each possible combo of weather and food values as groups
```

>> * <font color=DeepPink>Using the new groups:</font>
>>> * <font color=DeepPink>Use the `numpy` function `agg()` to **find:** <center>The mean and median number and price

```python
# .agg() function can return a scalar, Series, or DataFrame
# In this case, we are passing a list of DataFrame function names as the 'func=' argument,
# which returns a DataFrame
# Source: pandas - https://pandas.pydata.org/pandas-docs/stable/reference/api/
# pandas.DataFrame.agg.html

print("Aggregated mean and median of each Weather/Food group:\n", 
      weather_food.agg([np.mean, np.median]))
```

# 7. **Pandas DataFrames:**<br><center>Joining DataFrames

> <font color=deeppink>Using the two CSV files `dest.csv` and `tips.csv`, we will bring together two datasets, also known as a **join**.

```python
# Load the dest.csv data into a DataFrame and print the results
dest = pd.read_csv('https://github.com/AlanaDAg/'
                   'Python-Data-Analysis-Second-Edition/'
                   'blob/master/Chapter03/dest.csv?raw=True')
print("Dest:\n")
dest
```

```python
# Load the tips.csv data into a DataFrame and print the results
tips = pd.read_csv('https://github.com/AlanaDAg/'
                   'Python-Data-Analysis-Second-Edition/'
                   'blob/master/Chapter03/tips.csv?raw=True')
print("Tips:\n")
tips
```

<font color=deeppink>Using the `merge()` function, bring `dest` and `tips` together on the "EmpNr" column and **print** the results
> <font color=deeppink>Using the `join()` function, **query** both files and **print** the results.

```python
# Merge on 'EmpNr' column using merge() function
print("Files merged on column key 'EmpNr':\n", pd.merge(dest, tips, on='EmpNr'))

# Join with dest on the left and tips on the right
# Query using the .join() function
print("\nDest joined with Tips:\n", dest.join(tips, lsuffix='Dest', rsuffix='Tips'))
```
