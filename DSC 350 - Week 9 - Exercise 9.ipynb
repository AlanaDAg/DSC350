# <font color=blue>**2.**</font> **Storing CSV Files:** <center> Using NumPy and Pandas for Pickling & Unpickling


> **Generate** <font color=deeppink>a 365x4 NumPy array with random values.
> > <font color=deeppink>Then perform these steps:
> > * <font color=blue>**a.**</font> **Store** <font color=deeppink>the array in a CSV file and</font> **check** <font color=deeppink>its size
> > * <font color=blue>**b.**</font> **Save** <font color=deeppink>the array in the NumPy format,</font> **load** <font color=deeppink>the array,</font> **check** <font color=deeppink>its **shape** and the **size** of the file
> > * <font color=blue>**c.**</font> **Create a DataFrame** <font color=deeppink>from this array you have created and</font> **write it to a pickle**<font color=deeppink>, then</font> **retrieve** <font color=deeppink>it from the pickle
> > * <font color=blue>**d.**</font> **Print** <font color=deeppink>the size of the pickle

```python
# Import additionally necessary libraries
from tempfile import NamedTemporaryFile
from os.path import getsize

# Seed the random number generator
np.random.seed(50)

# Generate a 365x4 NumPy array
ex2_array = np.random.randn(365, 4)

# a. Store the array in a csv file and check its size
tmpf_2a = NamedTemporaryFile()
np.savetxt(tmpf_2a, ex2_array, delimiter=',')
print("Ex2 CSV file size", getsize(tmpf_2a.name))
```

```python
# b. Save the array in the NumPy format
tmpf_2a = NamedTemporaryFile()
np.save(tmpf_2a, ex2_array)
tmpf_2a.seek(0)   # simulates closing and reopening the temporary file

# Load the array
load_2b = np.load(tmpf_2a)

# Check its SHAPE and file SIZE 
print("2b array shape:\n", load_2b.shape)   # Shape
print("\n2b array .npy file size:\n\n", getsize(tmpf_2a.name))   # Size

# Print file name to see if this will help with Permissions Error ???
print(tmpf_2a.name)
```

```python
# Try importing to_pickle to fix Permissions Error ???
from pandas.io.pickle import to_pickle

# Create a DataFrame from this array and write it to a pickle
df_2b = pd.DataFrame(ex2_array)
#df_2b.to_pickle(tmpf_2a.name)   # Raises Permissions Error
df_2b.to_pickle(r"C:\Users\alana\AppData\Local\Temp\tmpjucsmz50.csv")   # Error

# Then retrieve it from the pickle
print("Retrieved DataFrame from pickle:\n", pd.read_pickle(tmpf_2a.name))

# d. Print the size of the pickle
print("\nPickled DataFrame size:\n", getsize(tmpf_2a.name))
```

### **Notes:**
> `NamedTemporaryFile()` **method:**
> * Creates a temporary file which temporarily stores data.
> * The temp file is automatically deleted once the file object is closed

> `getsize()` **function:**
> * A function that allows Python to interact with the operating system
> * Included in the Python OS module

> `seek()` **method:**
> * The seek() method does two things: <br>
>    **1.** sets the cvurrent file position in a file stream <br>
>    **2.** returns the new position
> * **Syntax**: *file*.seek(*offset*)
>    * *offset* parameter = Required. A number representing the position to set the current file stream position.

"EU","isp":"Tilaa V.O.F.","longitude":5.75,"latitude":52.5,"country_code": 
"NL","country_code3":"NLD"}' 

> <font color=deeppink>Using the above JSON string, <font color=black> **parse** <font color=deeppink>a JSON string with the `loads()` function, then:
> > * <font color=blue>**a.**</font> **Print** <font color=deeppink>the values for the "Country" column
> > * <font color=blue>**b.**</font> **Overwrite** <font color=deeppink>the value for Netherlands with a value of your choice
> > * <font color=blue>**c.**</font> **Print** <font color=deeppink>your results

```python
# Import necessary libraries
import json

# Set JSON string as Python object
json_string = '{"country":"Netherlands","dma_code":"0","timezone":"Europe/Amsterdam","area_code":"0","ip":"46.19.37.108","asn":"AS196752","continent_code":"EU","isp":"Tilaa V.O.F.","longitude":5.75,"latitude":52.5,"country_code":"NL","country_code3":"NLD"}'

# Parse JSON string with the loads() function
json_data = json.loads(json_string)

# a. Print the values for the "Country" column
print("Country column:\n", json_data["country"])
```

```python
# b. Overwrite the value for Netherlands with a value of your choice
json_data["country"] = "Finland"

# c. Print your results
print(json.dumps(json_data))
```

# <font color=blue>**5.**</font> **Creating & Converting File Types:** <center> To and From JSON Strings & Pandas Series


> <font color=deeppink>Using Pandas `read_json()` function, we can either create a Pandas Series or DataFrame. <br>
> Taking the JSON string from <font color=blue>#**4**<font color=deeppink>,<font color=black> **create** <font color=deeppink> a **Pandas Series**.

> > * <font color=blue>**b.**</font> **Convert** <font color=deeppink>the Pandas Series to a JSON string

```python
# Suppress FutureWarning statements about 'read_json' deprecation
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Create a Pandas Series with the JSON string from #4
json_data = pd.read_json(json_string, typ='series')
print("New Pandas Series data:\n", json_data)
```

```python
# a. Change the "Country" value again to a value of your choice
json_data["country"] = "Denmark"
print(json_data)
```

```python
# b. Convert the Pandas Series to a JSON string
pdseries_to_json = json_data.to_json()
print("Converted Pandas Series to JSON string:\n", pdseries_to_json)
```
