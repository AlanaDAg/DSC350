```python
# Import libraries
import pandas as pd
import numpy as np

# Use Pandas read_html() function to scrape HTML data from Wiki page
## This function returns a list of DataFrames --> it must be indexed
christie_dfs = pd.read_html('https://en.wikipedia.org/wiki/'
                            'Agatha_Christie_bibliography')[0]  # Index the return

# Find total number of tables on page
print("Total number of tables on Agatha Christie Wiki:\n", len(christie_dfs))

# Print the list of DF objects (tables) on Wiki page
print("\nTables on Christie's Wiki:\n", christie_dfs)
```

```python
# Find the specific "Christie's novels" table from the Wiki page
## Set as Pandas DataFrame
christie_noveltable = pd.read_html('https://en.wikipedia.org/wiki/'
                                   'Agatha_Christie_bibliography',
                                   match='Christie\'s novels')
                             # Use escape character backslash
                            ## allows Python to read the apostrophe in the string

# View the first five rows of "Christie's novels" table
display(christie_noveltable[0].head())
```

<font size=6><center>**Transformation 1**</font></center>
> <font size=4>Separate UK and US data <br> --- Store **UK** Data in a new DataFrame</font>

```python
# Store UK data in its own, separate DataFrame
## Column headers = "UK title[1][12][13]", "Year of UK publication", 
### "UK publisher (All London)", and "Series"

# Fetch column data
UK_Title = christie_noveltable[0]['UK title[1][12][13]']
Year_of_UK_Publication = christie_noveltable[0]['Year of UK publication']
UK_Publisher = christie_noveltable[0]['UK publisher (All London)']
Series = christie_noveltable[0]['Series']

# Create UK DataFrame
UK_data = pd.DataFrame({'UK title': UK_Title,
                        'Year of UK publication': Year_of_UK_Publication,
                        'UK publisher (All London)': UK_Publisher,
                        'Series': Series
                       })

# View the UK DataFrame
print("Agatha Christie Novels:\n UK data")
UK_data
```

<font size=6><center>**Transformation 2**</font>
> <font size=4>Rename New UK DataFrame Headers</font>

```python
# Rename column headers using dictionary structure and .rename() method
## Create dictionary of new headers
newheaders_UK = {'UK title': 'UK_Title',
                 'Year of UK publication': 'Year_of_UK_Publication',
                 'UK publisher (All London)': 'UK_Publisher'
                }

## .rename() method
UK_data.rename(columns=newheaders_UK,
               inplace=True)

# View UK DataFrame with transformed headers
print("Agatha Christie Novels:\n UK data (with transformed headers\n")
UK_data
```

<font size=6><center>**Transformation 3**</font>
> <font size=4>Add New '*Authors*' Column to UK DataFrame</font>

```python
# Add 'Authors' column to UK DataFrame

UK_data.insert(0, 'Authors', 'Agatha Christie')

# View the completed UK DataFrame with 'Authors' column and transformed headers
print("Agatha Christie Novels:\n Completed UK data\n")
UK_data
```

<font size=6><center>**Transformation 4**</font>
> <font size=4>Separate UK and US data <br> --- Store **US** Data in a new DataFrame</font></center>

```python
# Store US data in its own, separate DataFrame
## Column headers = "US title", "Year of US publication", 
### "US publisher (All New York), and "Series"

# Fetch column data
US_Title = christie_noveltable[0]['US title']
Year_of_US_Publication = christie_noveltable[0]['Year of US publication']
US_Publisher = christie_noveltable[0]['US publisher (All New York)']
# Series column data has already been fetched --> Series

US_data = pd.DataFrame({'US title': US_Title,
                        'Year of US publication': Year_of_US_Publication,
                        'US publisher (All New York)': US_Publisher,
                        'Series': Series
                       })

# View the US DataFrame
print("Agatha Christie Novels:\n US data")
US_data
```

<font size=6><center>**Transformation 5**</font>
> <font size=4>Rename New US DataFrame Headers</font>

```python
# Rename column headers using dictionary structure and .rename() method
## Create dictionary of new headers
newheaders_US = {'US title': 'US_Title',
                 'Year of US publication': 'Year_of_US_Publication',
                 'US publisher (All New York)': 'US_Publisher',
                }

## .rename() method
US_data.rename(columns=newheaders_US,
               inplace=True)

# View US DataFrame with transformed headers
print("Agatha Christie Novels:\n US data (with transformed headers)\n")
US_data
```

<font size=6><center>**Transformation 6**</font>
> <font size=4>Add New '*Authors*' Column to US DataFrame</font>

```python
# Add 'Authors' column to US DataFrame
US_data.insert(0, 'Authors', 'Agatha Christie')

# View the completed UK DataFrame with 'Authors' column and transformed headers
print("Agatha Christie Novels:\n Completed US data\n")
US_data
```
