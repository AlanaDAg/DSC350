# 1. **Subsetting Data** with indexing
#### <center>& **Descriptive Statistics:** Percentiles

<font color=DeepPink>Find the **95th percentile** of earthquake magnitude in Japan using the `mb` magnitude type.

> **REMEMBER:** The 95th percentile is the 95th place out of 100 places. So this question is asking you, "What earthquake magnitude in Japan (using the mb magnitude type) comes in 95th place out of 
<center>100% - 5% = 95% <br>1 - 0.05 = 0.95

```python
# Importing Pandas was performed in the Prep section
# Find the columns of concern
## Japan in 'parsed_place' column; `mb` magnitude type in 'magType' column
## Include earthquake title

df.set_index('parsed_place').filter(like='Japan', axis=0).filter(items=
                                                            ['mag', 'magType', 'title'])

# The 'parsed_place' column is the index 
# The .filter() method inputs arguments (can be stacked)
## like= for things present in the specified axis (0 is the first column in a DataFrame)
## items= to specify the other features you want displayed
```

```python
# Find the 95th percentile of earthquake magnitude in Japan
## Specifically of the `mb` magnitude type (column 'magType')

percentile = df[(df.parsed_place == 'Japan') & (df.magType == 'mb')].mag.quantile(0.95)

print("The 95th percentile of earthquake magnitude in Japan of the `mb` magnitude type" +
      f" is: \n{percentile}")
```

# 2. **Descriptive Statistics:** Percentages

<font color=DeepPink>Find the **percentage** of earthquakes in Indonesia that were **coupled** with tsunamis.

```python
# Find the columns of concern
## Indonesia in 'parsed_place' column; 'tsunami' along with earthquake-related columns

df.set_index('parsed_place').filter(like='Indonesia', axis=0).filter(items=
                                          ['mag', 'magType', 'title', 'tsunami', 'type'])
```

```python
# Find the number of rows with 'tsunami'
df.tsunami.value_counts()
```

```python
# Create object to find the number of rows with Indonesia and
## number of rows with tsunamis
### Use the .value_counts() function to find specified values
### Use the .iloc() method to find the tsunami values in the 'Indonesia' index
decimal = df[df.parsed_place == 'Indonesia'].tsunami.value_counts(normalize=True).iloc[1,]
#print(decimal)

# Multiply decimal by 100 and round to two decimal places
percentage = round((decimal * 100), 2)
print("The percentage of earthquakes in Indonesia that were coupled with tsunamis is:" +
      f" \n{percentage}%")
```

# 4. **Pandas Dataframe:** Adding Columns

<font color=DeepPink>**Add a column** indicating whether the earthquake happened in a country or US state that is on the Ring of Fire.
>Use:
>> Alaska, Antarctica (look for Antarctic), Bolivia, Claifornia, Canada, Chile, Costa Rica, Ecuador, Fiji, Guatemala, Indonesia, Japan, Kermadec Islands, Mexico (be careful not to select *New* Mexico), New Zealand, Peru, Philippines, Russia, Taiwan, Tonga, and Washington.

```python
# Find the correct Mexico label in the 'parsed_place' column
# Sort the 'parsed_place' column alphabetically and by unique values
df.parsed_place.sort_values().unique()
```

```python
# Add a new column to the DataFrame
## Use .join() to change the tuple [] into a string of characters
df['Ring_of_Fire'] = df.parsed_place.str.contains(r'|'.join([
    'Alaska', 'Antarctica', 'Bolivia', 'California', 'Canada', 'Chile',
    'Costa Rica', 'Ecuador', 'Fiji', 'Guatemala', 'Indonesia', 'Japan',
    'Kermadec Islands', 'Mexico', 'New Zealand', 'Peru', 'Philippines',
    'Russia', 'Taiwan', 'Tonga', 'Washington']))
```

```python
# See 'parsed_place' column and new 'ring_of_fire' column
df[["parsed_place", "Ring_of_Fire"]]
```

# 6. **New DataFrame Column:** Tsunami Calculations

<font color=DeepPink>Find the **tsunami** count along the Ring of Fire.
