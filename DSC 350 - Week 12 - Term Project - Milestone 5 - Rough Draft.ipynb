```python
# SQLite3

# Import libraries
import csv
import sqlite3

# Establish connection to database
with sqlite3.connect(":memory:") as con:
    cursor = con.cursor()
    flatfile_table = '''CREATE TABLE Agatha_Christie_Novels(
                        Authors INTEGER PRIMARY KEY,
                        ID INTEGER,
                        book_ID INTEGER NOT NULL,
                        best_book_id INTEGER NOT NULL,
                        work_ID INTEGER NOT NULL,
                        books_count INTEGER NOT NULL,
                        ISBN INTERGER NOT NULL,
                        ISBN_13 INTEGER NOT NULL,
                        OG_publication_year INTEGER NOT NULL,
                        OG_title TEXT NOT NULL,
                        title TEXT NOT NULL,
                        lang_code TEXT NOT NULL,
                        mean_rating INTEGER NOT NULL,
                        ratings_count INTEGER NOT NULL,
                        work_rarings_count INTEGER NOT NULL,
                        work_txtreviews_count INTEGER NOT NULL,
                        rate_1 INTEGER NOT NULL,
                        rate_2 INTEGER NOT NULL,
                        rate_3 INTEGER NOT NULL,
                        rate_4 INTEGER NOT NULL,
                        rate_5 INTEGER NOT NULL,
                        image_url TEXT NOT NULL,
                        sml_img_URL TEXT NOT NULL);
                        '''
    cursor.execute(flatfile_table)

    file = open(r'C:\Users\alana\OneDrive\Desktop\DSC 350\FlatFile_ Term_Project_Data/DSC 350 - Milestone 5 - books flat file data.csv')

    contents = csv.reader(file)

    insert_records = "INSERT INTO Agatha_Christie_Novels (ID, book_ID, best_book_id, work_ID, books_count, ISBN, ISBN_13, OG_publication_year, OG_title, title, lang_code, mean_rating, ratings_count, work_rarings_count, work_txtreviews_count, rate_1, rate_2, rate_3, rate_4, rate_5, image_url, sml_img_URL) Values(?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)"
    cursor.executemany(insert_records, contents)
    
                        
    
```
